```python
from google_play_scraper import Sort, reviews_all, reviews
import pandas as pd
import numpy as np
```

```python
result, continuation_token = reviews(
    'com.shopee.id',
    lang='id', #bahasa
    country='id', #negara
    sort=Sort.MOST_RELEVANT, #sorting yg paling relevan
    count=5000, #jumlah dataset
    filter_score_with= None  #Isi dengan 1, 2, 3, 4, 5 None jika ingin bercampur
)
```

```python
# Dataframe dengan nama
data = pd.DataFrame(np.array(result),columns=['review'])
data = data.join(pd.DataFrame(data.pop('review').tolist()))
data.head()
```

```python
data.to_csv("Ulasan Shopee 5000 Data.csv", index = False  , encoding='utf-8') #menyimpan dataset dalam bentuk csv
```
